```python
c1=pd.read_csv(r'D:\notebook\data20200206\contb_01.csv')
c2=pd.read_csv(r'D:\notebook\data20200206\contb_02.csv')
c3=pd.read_csv(r'D:\notebook\data20200206\contb_03.csv')
df=pd.concat([c1,c2,c3],ignore_index=True)
df
```

-    2、***数据清洗***
-    2.1、缺失值处理
-    2.1.1、查看带有缺失值的列
-    2.1.2、空值指定填充

```python
# 填充空值，NOT PROVIDE没有提供
df['contbr_employer'].fillna(value='NOT PROVIDE', inplace=True) #用NOT PROVIDE填充空值 说明这些值没有被提供
#填充contbr_occupation为空的值为NOT PROVIDE
df['contbr_occupation'].fillna('NOT PROVIDE', inplace=True)
df['contbr_st'].fillna('NOT PROVIDE', inplace=True)
df.info() # 查看填充后的结果
```

```python
#同时我们发现很多捐赠金额是负数，虽然没有缺失值但是异常值需要删除用布尔值处理就行了
df['contb_receipt_amt']<=0
df=df[~(df['contb_receipt_amt']<=0)]
df.info()
```

- 2.2.1、候选人去重
- 2.2.2、添加党派

```python
#2.2.1  去重可以知道大概有13个候选人
df.cand_nm.unique() 
```

```python
#2.22添加党派 建立映射字典并加入映射函数map处理
parties = {
  'Bachmann, Michelle': 'Republican',
  'Romney, Mitt': 'Republican',
  'Obama, Barack': 'Democrat',
  "Roemer, Charles E. 'Buddy' III": 'Republican',
  'Pawlenty, Timothy': 'Republican',
  'Johnson, Gary Earl': 'Republican',
  'Paul, Ron': 'Republican',
  'Santorum, Rick': 'Republican',
  'Cain, Herman': 'Republican',
  'Gingrich, Newt': 'Republican',
  'McCotter, Thaddeus G': 'Republican',
  'Huntsman, Jon': 'Republican',
  'Perry, Rick': 'Republican'           
 }
df['party']=df['cand_nm'].map(parties)
df
```

```python
df.groupby('party').count()
```

```python
#2.2.3统计两个党派支持的次数
df.party.value_counts()
```

- 可以发现民主党支持率明显高很多，很可能会赢得大选

```python
#2.2.4、查询每个政党的捐献额
df.groupby('party')['contb_receipt_amt'].sum()
```

```python
#2.2.5、按照职业汇总对赞助总金额进行排序
df.groupby('contbr_occupation')['contb_receipt_amt'].sum().sort_values(ascending=False).head(50)
```

```python
# #2.2.6、职业类型去重
#整理一部分相同的职业，如果全部整理需要花费很长时间
occupation = {'INFORMATION REQUESTED PER BEST EFFORTS':'NOT PROVIDE',
             'INFORMATION REQUESTED':'NOT PROVIDE',
              'C.E.O.':'EXECUTIVE',
              'DIRECTOR':'EXECUTIVE',
              'PRESIDENT':'EXECUTIVE',
              'VICE PRESIDENT':'EXECUTIVE',
              'PRESIDENT & C.E.O':'EXECUTIVE',
              'LAWYER':'ATTORNEY',
              'SELF':'SELF-EMPLOYED',
              'SELF EMPLOYED ':'SELF-EMPLOYED'}
f = lambda x : occupation.get(x,x)
#进行map转换
df['contbr_occupation'] = df['contbr_occupation'].map(f)
df['contbr_occupation']
```

```python
df.groupby('contbr_employer')['contb_receipt_amt'].sum().sort_values(ascending=False).head(50)
```

```python
employer = {'INFORMATION REQUESTED PER BEST EFFORTS':'NOT PROVIDE',
            'INFORMATION REQUESTED (BEST EFFORTS)':'NOT PROVIDE',
            'INFORMATION REQUESTED':'NOT PROVIDE',
            'NONE':'NOT PROVIDE',
            'REQUESTED ':'NOT PROVIDE',
            'SELF':'SELF-EMPLOYED',
            'SELF EMPLOYED':'SELF-EMPLOYED'}
g= lambda x : employer.get(x,x)
df['contbr_employer'] = df['contbr_employer'].map(g)
df['contbr_employer']
```

- 3.1、透视表(pivot_table)分析党派和职业

```python
df_p_o=df.pivot_table(index='contbr_occupation',columns='party',values='contb_receipt_amt',aggfunc=sum,fill_value=0)
```

```python
#计算出每个职业给两党献金的总额，并排除总额小于200w的人数
df_p_o['total']=df_p_o['Democrat']+df_p_o['Republican'] 
df_p_o[df_p_o['total']>=2000000].applymap(lambda x: '% .2f' % x)#将所有数据改成正常计数而不是科学计数
```

#3.2、分组及运算和转换
- 3.2.1、不同职业对他两人的支持度
- 3.2.2、不同公司对他两人的支持度

```python
df_cn_o=df.pivot_table(index='contbr_occupation',columns='cand_nm',values='contb_receipt_amt',aggfunc=sum,fill_value=0)
df_cn_o[['Obama, Barack', 'Romney, Mitt']]
```

```python
obama_amount=df_cn_o['Obama, Barack']>df_cn_o['Romney, Mitt'] #奥巴马比罗姆尼献金总数高的职业数据表
romney_amount=df_cn_o['Romney, Mitt']>df_cn_o['Obama, Barack'] #罗姆尼比奥巴马献金总数高职业数据表
df_cn_o['Obama, Barack'][obama_amount].sort_values(ascending=False).head(10)
```

- 可以看出 支持奥巴马政府 主要来自退休人员，律师，医生，咨询师，企业CEO，失业人士 

```python
df_cn_o['Romney, Mitt'][romney_amount].sort_values(ascending=False).head(10)
```

- 可以看支持罗姆尼的 来自家庭妇女，公司主管，房地产，银行投行，投资者，自由职业者 
两者支持的群体有很大差别

```python
df_cn_em=df.pivot_table(index='contbr_employer',columns='cand_nm',values='contb_receipt_amt',aggfunc=sum,fill_value=0)
df_cn_em.head(5)
```

```python
#3.2.2、不同公司对他两人的支持度
obama_amount1=df_cn_em['Obama, Barack']>df_cn_em['Romney, Mitt'] #奥巴马比罗姆尼献金总数高的职业数据表
romney_amount1=df_cn_em['Romney, Mitt']>df_cn_em['Obama, Barack'] #罗姆尼比奥巴马献金总数高职业数据表
df_cn_em['Obama, Barack'][obama_amount1].sort_values(ascending=False).head(15)
```

- 同理也可以得出支持奥巴马的大多是实业资本，支持罗姆尼的大多是金融资本

- 3.3.2、绘制Obama和Romney各区间赞助的总金额

```python
df.groupby(['cand_nm','label'])['contb_receipt_amt'].size().unstack(fill_value=0).T[['Obama, Barack','Romney, Mitt']]
```

```python
obama_romney_amount=df.groupby(['cand_nm','label'])['contb_receipt_amt'].size().unstack(fill_value=0).T[['Obama, Barack','Romney, Mitt']]
obama_romney_amount.plot.bar(figsize=(8,4))
```

- 大金额干扰比较大，可以考虑排除

```python
obama_romney_amount[:-2].plot.bar(figsize=(8,4))
```

```python
# 百分比堆积图
obama_romney_amount.div(obama_romney_amount.sum(1), axis=0)[:-2].plot(kind='bar', stacked=True, figsize=(10, 6))
plt.legend(loc='center')
```

```python
obama_romney_amount.div(obama_romney_amount.sum(1),axis=0) #axis指明运算方向。这里面sum有一个运算方向是1，div本身的运算方向是0
```
