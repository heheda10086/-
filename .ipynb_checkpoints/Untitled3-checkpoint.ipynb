### 1数据提取融合和预处理

```python
c1=pd.read_csv(r'D:\notebook\data20200206\contb_01.csv')
c2=pd.read_csv(r'D:\notebook\data20200206\contb_02.csv')
c3=pd.read_csv(r'D:\notebook\data20200206\contb_03.csv')
df=pd.concat([c1,c2,c3],ignore_index=True)
df
```

-    2、***数据清洗***
-    2.1、缺失值处理
-    2.1.1、查看带有缺失值的列
-    2.1.2、空值指定填充

```python
# 填充空值，NOT PROVIDE没有提供
df['contbr_employer'].fillna(value='NOT PROVIDE', inplace=True) #用NOT PROVIDE填充空值 说明这些值没有被提供
#填充contbr_occupation为空的值为NOT PROVIDE
df['contbr_occupation'].fillna('NOT PROVIDE', inplace=True)
df['contbr_st'].fillna('NOT PROVIDE', inplace=True)
df.info() # 查看填充后的结果
```

```python
#同时我们发现很多捐赠金额是负数，虽然没有缺失值但是异常值需要删除用布尔值处理就行了
df['contb_receipt_amt']<=0
df=df[~(df['contb_receipt_amt']<=0)]
df.info()
```

- 2.3.4、选取选举人为Obama、Romney的子集数据

```python

#方法1 query取范围
df.query("cand_nm == 'Obama, Barack' or cand_nm == 'Romney, Mitt'")
```

```python
#方法2 带入一个与运算
c1 = df['cand_nm'] == 'Obama, Barack'
c2 = df['cand_nm'] == 'Romney, Mitt'
# 这是一个与运算
cond = c1|c2

df_or = df[cond]
df_or
```

```python
#方法3 isin 带入含有两者的列表 求出布尔值，然后布尔值代入df
oandr= df['cand_nm'].isin(['Obama, Barack', 'Romney, Mitt'])
df[oandr]
```
