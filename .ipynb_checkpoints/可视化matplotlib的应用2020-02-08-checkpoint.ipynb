```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif']=['SimHei']
plt.rcParams['axes.unicode_minus']=False #显示中文和正负号
%config InlineBackend.figure_format='svg' #增加清晰度
```

#1必须掌握的图 plt.plot 线形图

```python
x=[1,2,3,4,5,6,7]
y=[21,5,56,42,345,21,3]
plt.plot(x,y)
```

```python
x=np.linspace(-np.pi,np.pi,30)
y=x**2
plt.plot(x,y, label='天气')
plt.plot(x*2,y-3,label='心情')
plt.legend()
```

```python
fig=plt.figure()  
#1先实例化一个图像
x=np.linspace(-np.pi,np.pi,30)
y=x**2
plt.plot(x,y)
plt.xlabel('年龄')
plt.ylabel('体重')
plt.title('其实并没有关系是你吃的太多了')
#保存图像 保存到当下目录。
fig.savefig('./123.png')
```
