pandas 中的nan和none
- 请问下列随机数组中，将没有空值的行取出来

```python
np.random.seed(10)#加上时间种子锁定固定模式
df=DataFrame(np.random.randint(0,100,size=(8,6))) 
df
```

```python
df.iloc[2,3]=np.nan #位置索引赋值，iloc是位置索引
df.iloc[5,5]=None  #这里会默认强行转成float浮点型的nan值方便计算
df.iloc[5,3]=np.nan
df.iloc[7,2]=np.nan
df
```

数据清洗填充，删除

```python
#判定空值
df.isnull()
```

```python
df.isnull().any(axis=1)#any只要有一个就返回true 从左到右，只要有一个是null就返回true axis是指方向1 结果显示是  是空的值
```

```python
df.notnull().all(axis=1)#为什么上面用any 下面用all 下面说 notnull用 都不是null第一行都不是null第三行有null
#用false 取非空  结果显示不是空的值，和题目要求一致不需要取反
```

```python
df.loc[df.notnull().all(axis=1)]#取非空的行索引进行普通索引，
```

```python
df[~(df.isnull().any(axis=1))]#根据题目要求要取反
```

# isnull or notnull都很麻烦，有直接过滤空值的办法dropna
 - dropna就是直接去除带有空值的行

```python
df.dropna(axis=0)#dropna效果和取notnull().运算方向向下，所以是axis=0，axis和运算方向一致
```

#一般不用df。dropna（axis=1）因为如果用inplace=true容易替换原有表格

#空值的填充 fillna（）

```python
df.fillna(value=686)
```

### 用周围的值填充空值，增加可读性 

```python
df.fillna(method='ffill',axis=0)#axis是方向，我们是逐一竖直方向填充，用axis=0，
```

### 有的数据填充了发现第一行还是nan是因为第一行没有上一行无法填充那就就地加一个bfill就地填充下一行这样就没空值了

```python
df.fillna(method='bfill',axis=1)#一个是向前填充，一个是向后填充，为了数据更贴近数据 forward，behind
```

#检测空值是否填充完毕 需要检测列，因为列是主要索引，检测行数据太多了

    #处理重复值 drop_duplicates只能删除行，因为没有axis选项也就是默认axis是固定的axis=0 向下drop

```python
df1=DataFrame(np.random.randint(0,100,size=(10,8)))
df1.iloc[1]=[1,1,1,1,1,1,1,1]
df1.iloc[3]=[1,1,1,1,1,1,1,1]
df1.iloc[5]=[1,1,1,1,1,1,1,1]
df1.iloc[7]=[1,1,1,1,1,1,1,1]
df1
```

```python
df1.drop_duplicates(keep="first") #first就是保留第一个 last就是保留最后一个重复的行 默认是去掉重复行
```

#### 自定义一个df，区间在0-1之间，1000行三列，取值范围0-1之间 abc列，其中c列中的数值大于其两倍标准差
####  的值就算是异常值，并对异常值进行清洗

```python
np.random.seed(10)
df2=DataFrame(np.random.random(size=(1000,3)),columns=["A","B","C"])
df2.head()
```

```python
std_1=df2['C'].std()*2
std_1# 求出c的两倍标准差
```

```python
~(df2['C']>std_1).head() #~取反的符号
```

```python
df2.loc[~(df2['C']>std_1)]# 用的是标签索引是普通索引，要用loc来表示出来
```
