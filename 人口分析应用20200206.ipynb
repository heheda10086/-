```python
#6合并各州面积数据area pop 和area融合找共同项 因为简称的关系，所以要合并三个表
abb_pop_area=pd.merge(abb_pop,area,on="state",how="outer")
abb_pop_area.head()
```

```python
#7我们会发现area（sq mi）有缺失值，请找出并去除含有缺失值得行
abb_pop_area['area (sq. mi)'].isnull()
drop_index=abb_pop_area.loc[abb_pop_area['area (sq. mi)'].isnull()].index
```

```python
abb_pop_area.drop(labels=drop_index,axis=0,inplace=True)
```

```python
#8找出2010年全民人口数据 
abb_pop_area.query('ages=="total" & year == 2010')
```

```python
#9计算各州的人口密度 #人数/面积
abb_pop_area["midu"]=abb_pop_area["population"]/abb_pop_area["area (sq. mi)"]
abb_pop_area.head()
```

```python
#10排序并找出人口密度最高的五个州 #sort_values?
df1=abb_pop_area.sort_values(by="midu",ascending=False)
abb_pop_area.sort_values(by="midu",ascending=False)['state'].unique()[0:5].tolist()
```

```python
#改成2010年的人口密度排序
df1['year']==2010
df2=df1.loc[df1['year']==2010]
df2.groupby('state').max().sort_values(by="midu",ascending=False)
```
