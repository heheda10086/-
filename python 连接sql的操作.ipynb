```python
import pymysql
import pandas as pd
#pip install pymysql
```

```python

eng=pymysql.connect(host='localhost',
                   user='root',
                    password='123456',
                   db='test1',
                   charset='utf8')
```

```python
course='SELECT * FROM course'
df1=pd.read_sql(course,eng)
df1
```

```python
sc='SELECT * FROM sc'
df2=pd.read_sql(sc,eng)
df2
```

```python
df3=pd.merge(df1,df2,on="cid")
df3
```

```python
student='SELECT * FROM student'
df_student=pd.read_sql(student,eng)
df_student
```

```python
teacher='SELECT * FROM teacher'
df_teacher=pd.read_sql(teacher,eng)
df_teacher
```

```python
df4=pd.merge(df3,df_teacher,on='tid')
df4
```

```python
df_all=pd.merge(df4,df_student, on='sid',how='outer')
```

```python
df_all.sort_values(by='sid')
```
