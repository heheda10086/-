## query的应用 apply函数的应用

```python
def aggfunc(x,y):
    return df3.query('close>@x  and volume>@y') #因为直接饮用x，y会直接认为没有定义 所以这里用到@x调用外来自变量x，y

aggfunc(400,40000)#调用含范围的函数 用取值来限定需要的条件项
# 方法扩展，query 取范围，从df中取query范围，不需要切片[]来取直接外用引号限定要求就好了 如上的案例
```

```python
def fun(x):
    return x*3-30
df3['close'].apply(fun)  #对每一个函数，进行x*3-30的运算并返回结果，并对df3【close】的收盘值传入这个函数，也就是apply
                        #对每个序列的值进行函数操作并返回操作结果
```

## 双均线策略的制定

```python
df=ts.get_k_data('600023',start='2019-01-01')
df.head()
```

```python
#5日均线 30日均线  rolling函数的应用就是将列表中选取序列中的数值进行滚动操作，window取窗口的数量，这里是五日均线，那么就是
#rolling（5）选出五个为一组的序列进行mean（）均值运算求出需要的条目，一般rolling用于时间滚动操作
df['ma5']=df['close'].rolling(5).mean()
df['ma30']=df['close'].rolling(30).mean()
df.tail()
```

## 如果我从2010年1月1日开始，金叉买死叉卖，初始10w，到现在为止利润为多少

```python
#涉及了时间，做数据的预处理
df['date']=pd.to_datetime(df['date'])
```

```python
df1=df.set_index(['date']) ## df.set_index(df['date'],inplace=True)太麻烦还要drop掉多余的date列
```

```python
df1.head()
```

```python
a=df1[0:260]
plt.plot(a.index,a['ma5'],c='red')
plt.plot(a.index,a['ma30'],c='green')
```

```python
t1=df1['ma5']<df1['ma30']
#True说明ma5小于ma30是金叉前沿
#False说明ma5大于ma30是金叉后延 两者连续出现就说明是金叉，相反就是死叉
```

```python
t2=df1['ma5']>=df1['ma30']
```

- t1            T   F    F 》 T   T 》 F        ##F--T 死叉   T--F金叉
- t2.shift(1)         F    T   T   F   F   T
                 F    F   T   F   F    ## 这边的  T对应的值就是死叉交点对应的值，也就是说t1加上t2位移形成的布尔值就是死叉的点

```python
death=df1.loc[t1 & t2.shift(1)].index
death #死叉的点
```

```python
gold=df1.loc[~(t1 | t2.shift(1))].index
gold
```

```python
firstmoney=100000
money=100000
hold=0
s1=Series(1,index=gold)
s2=Series(0,index=death)
```

```python
s=s1.append(s2)
```

## 根据时间排序

```python
s.sort_index()
```

```python
#使用开盘价买
for i in range(0,len(s)):
    p=df1['open'][s.index[i]] 
    
    if s[i]==1:  #如果是金叉，等于是s【i】==1取值取到1是金叉买点
        buy=money//p*100 #整除买入多少手股票  钱//整除100股的价值
        hold=buy*100  #持有的股票数量
        money-=buy*100*p #把本金10w减去
    else :
        money+=hold*p  #卖股票 s【i】==0的状况 需要把卖剩下的钱加满
        hold=0  #卖光了手里没有股票
money+=hold*df1['open'][-1]
result=money-firstmoney
print(result)
```
