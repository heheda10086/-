#映射操作

```python
dic={'name':['张三','李四','王老五'],
   'salary':[22222,7777,132141]}
```

```python
df=DataFrame(dic)
df
```

```python
#建立映射关系表
dic={'张三':'jack',
     '李四':'henry',
     '王老五':'kate'}
#映射关系表
```

```python
df['name'].map(dic) #map指明了映射对象 意思就是将df['name']序列中的值代入map映射中得出对应的值（当然可以将新生成的值编辑成列）
```

- 将新生成的值编辑成列进行一一对应，可以用到很多场景，比如每个学生对应的老师映射，学生对应的班级映射，员工对应的部门经理的映射

```python
df['ename']=df['name'].map(dic)#可以给映射命名 查一下百度了解映射的用法
df
```

map的运算，3000以上部分抽取50%缴纳个税，计算每个人的税后薪资

```python
def after_salary(s):
    return s-(s-3000)*0.5
df['after_salary']=df['salary'].map(after_salary) #给map传一个函数，函数的参数前置也就是df【‘salary、’】
df
```

```python
#可以直接用lambda解决
df['salary'].map(lambda x: x-(x-3000)*0.5) #通过传入匿名函数直接省略了建立自定义函数的步骤也是得出一样的结果
```

## 排序实现的随机抽样 应用场景，比如不同部门的人抽验，或者不同产品抽验批次都是可以应用的
#### take()
#### np.random.permutation()

```python
np.random.seed(10)
df=DataFrame(np.random.randint(0,100,size=(100,3)),columns=["A","B","C"])
df.head()
```

```python
df.take([1,0,2],axis=1)#打乱排序，这里1,20只能用隐士列索引而不是BCA这样的显示索引 用显示索引会出错
#axis=1指明横轴方向，和顺序变动方向一致，所以是调整列索引，
#axis=0 则是调整行索引 和take的方向一致所以用0
```

```python
df.take(np.random.permutation(3),axis=1)[0:6] #np.random.permutation()打乱相应的行和列进行重新排序取样，行索可以不用loc
```

```python
df.take(np.random.permutation(100),axis=0).iloc[0:15,0:2]#打乱抽样，用位置索引解决
```

```python
df.take(np.random.permutation(3),axis=1).take(np.random.permutation(100),axis=0).iloc[0:15,0:2]#先列后行的打乱次序
```

```python
df=DataFrame({'item':["apple",'banana','orange','banana','orange','apple'],
             'price':[4,3,3,2.5,4,2],
             'color':['red','yellow','yellow','green','green','green'],
             'weight':[12,20,50,30,20,44]})
df
```

```python
df.groupby(by='item').groups
```

```python
#求各类水果平均价格
df.groupby(by='item').mean()['price']
```

```python
#计算各种颜色水果的平均重量
df.groupby(by='color').mean()['weight']
```

```python
#把每种水果的平均价格汇总到表格当中
df_series=df.groupby(by='item').mean()['price']
```

```python
df['mean_prcie']=df_series  #因为行数不匹配直接赋值会出错
df
```

# 可以转字典 类型的映射，避免手打的麻烦直接to_dict（操作）

```python
dic=df_series.to_dict() #可以转字典 类型的映射
dic
```

```python
df['mean_prcie']=df['item'].map(dic)#用map传入函数
df
```

```python
df.groupby(by='item')['price'].mean()
```

```python
def abc(s):
    return s
```

```python
df.groupby(by='item')['price'].transform(abc)
```

```python
#我想自定义一个求均值的函数，使其作用于分组结果之中
def mean1(s):
    print(s)
    sum=0
    for i in s:
        sum+=i
    return sum/s.size
df.groupby(by='item')['price'].transform(mean1)
```

```python
df.pivot_table(index=['item'],values=['price','color','weight'],aggfunc={'price':'mean'})
```

```python
#我想自定义一个求均值的函数，使其作用于分组结果之中，apply的结果 虽然很多余直接用mean聚合函数就可以解决问题了。主要是看函数的思路
def mean1(s):
    sum=0
    for i in s:
        sum+=i
    return sum/s.size
df.groupby(by='item')['price'].apply(mean1)
```
