```python
mf=ts.get_k_data('601318',start='2015-01-01')
mf
```

```python
type(mf)
```

```python
s=mf['close']
mf['ma5']=s.rolling(5).mean()
mf['ma30']=s.rolling(30).mean()
mf
```

```python
mf1=mf[:260]
```

```python
plt.plot(mf1.index,mf1['ma5'],c='red')
plt.plot(mf1.index,mf1['ma30'],c='blue')
```

```python
mf['date']=pd.to_datetime(mf['date'])
```

```python
mf.set_index(mf['date'],inplace=True)
mf.drop(labels='date',axis=1,inplace=True)
mf
```

```python
gold0=mf['ma5']<mf['ma30']
death0=mf['ma5']>=mf['ma30']
```

```python
death1=mf.loc[gold0&death0.shift(1)].index
death1
```

```python
gold1=mf.loc[~(gold0|death0.shift(1))].index
gold1
```

```python
firstmoney=100000
money=firstmoney
s1=Series(1,index=gold1)
s2=Series(0,index=death1)
s=s1.append(s2) #拼接字段
s=s.sort_index()  
```

```python
for i in range(0,len(s)):
    p=mf['open'][s.index[i]]
    if s[i]==1:
        buy=money//p*100
        hold=buy*100
        money-=buy*100*p
    else :
        money+=hold*p
        hold=0
money+=hold*mf['open'][-1]
result=money-firstmoney
print(result)
```
