pandas 中的nan和none

```python
df=DataFrame(np.random.randint(0,100,size=(8,6))) 
df
```

```python
df.iloc[2,3]=np.nan #位置索引赋值，iloc是位置索引
df.iloc[5,5]=None  #这里会默认强行转成float浮点型的nan值方便计算
df.iloc[5,3]=np.nan
df.iloc[7,2]=np.nan
df
```

数据清洗填充，删除

```python
#判定空值
df.isnull()
```

```python
df.isnull().any(axis=1)#any只要有一个就返回true 从左到右，只要有一个是null就返回true 
```

```python
df.notnull().all(axis=1)#为什么上面用any 下面用all 下面说 notnull用 都不是null第一行都不是null第三行有null
#用false 取非空
```

```python
df.loc[df.notnull().all(axis=1)]#取非空的行索引进行普通索引，
```

#isnull or notnull都很麻烦，有直接过滤空值的办法dropna

```python
df.dropna(axis=0)#dropna只能去掉有空值的 “ 行”  只能是行，效果和取notnull().all(axis=1)是一样的
```

#一般不用df。dropna（axis=1）因为如果用inplace=true容易替换原有表格

#空值的填充 fillna（）

```python
df.fillna(value=999)
```

```python
df.fillna(method='ffill',axis=0)#axis是方向，我们是逐一竖直方向填充，用axis=0，
```

```python
df.fillna(method='bfill',axis=1)#一个是向前填充，一个是向后填充，为了数据更贴近数据 forward，behind
```
